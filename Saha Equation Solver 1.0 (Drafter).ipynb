```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.optimize import root_scalar

from sympy import *
ionisation_lvls_eV = [[0.0, 13.598434599702], [0.0, 24.587389011, 54.417765486]]
m_e_eV = 510998.95000

def lam3(t_eV):
    return (2.0*np.pi/(m_e_eV*t_eV))**1.5

nE = symbols('n_e')
nH = 1
M = Matrix([[1,1,-nH], [0,1,-nE], [np.exp(-ionisation_lvls_eV[0][1]/1.0e3)/lam3(1.0e3),-nE,0]])
#M = Matrix([[0,1,0,1,2,-nE], [1,1,0,0,0,-nH], ...] 

expr = M.det()

print(expr)

def eq(ne):
    return expr.subs(nE,ne)

sol = root_scalar(eq, method='brentq', bracket=(0,1.0e6))
print(sol)
```

```python
#Solving for hydrogen by a quadratic equation

import cmath

T = 6700;
k = 8.617333*pow(10,-5);
n = pow(10,23);
C = 1.3018*pow(10,47);

a = 1
b = 0.5*C*pow(T,1.5)*exp(13.6/(k*T))
c = -0.5*n*C*pow(T,1.5)*exp(13.6/(k*T))

# calculating the discriminant
dis = (b**2) - (4 * a*c)

# find two results
ans1 = (-b-cmath.sqrt(dis))/(2 * a)
ans2 = (-b + cmath.sqrt(dis))/(2 * a)

# printing the results
print('The roots are')
print(ans1)
print(ans2)

```

```python
#Same as above but looping it

import cmath
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import root_scalar
from sympy import *

#Temp = [*range(1000,5000,1)]
Temp = 10**np.linspace(3,6,200)
Densities = [pow(10,23)];

k = 8.617333*pow(10,-5);
#C = 1.3018*pow(10,47);
C = 1;

n1H = [];

for T in Temp:
    for n in Densities:
        a = 1
        b = 0.5*C*pow(T,1.5)*exp(13.6/(k*T))
        c = -0.5*n*C*pow(T,1.5)*exp(13.6/(k*T))
        
        # calculating the discriminant
        dis = (b**2) - (4 * a*c)
        
        # find two results
        ans1 = (-b-cmath.sqrt(dis))/(2 * a)
        ans2 = (-b + cmath.sqrt(dis))/(2 * a)
        
        # printing the results
        print('The roots are')
        print(ans1)
        print(ans2)
        n1H.append(ans1)
        
n0H = np.copy(n1H); n0H[:] = [-x + pow(10,23) for x in n0H];
```
